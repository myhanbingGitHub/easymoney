```python
import pandas as pd
import tushare as ts
from sqlalchemy import create_engine
import time
import datetime
import json
import numpy as np
```

```python
# 导入token
with open('./parameters.json','r') as f:
    p=json.load(f)
    user = p['user']
    port = p['port']
    psw = p['password']
    host = p['host']
    token=p['TU_share_pro_taken']
    cnnstr = "mysql://" + user + ":" + psw + "@" + host + ":" + str(port) + "/stock?charset=utf8&use_unicode=1"
engine_ts = create_engine(cnnstr)
ts.set_token(token)
```

```python
# 设置参数
stocklist=['000001.SZ']
startdate='20110101'
enddate='20201031'
startmonth='201101'
endmonth='202010'
startQ='2011Q1'
endQ='2020Q4'
index=['000001.SH','000005.SH','000006.SH','000300.SH','000905.SH','399001.SZ','399005.SZ','399006.SZ','399016.SZ','399300.SZ','399905.SZ']
```

```python
# 获取股票历史
stock = ts.pro_bar(ts_code=stocklist[0], start_date=startdate, end_date=enddate,
                            asset='E', adj='qfq', freq='D')
stock.head(5)
```

```python
# 初始化合并用的数据框
combination=stock.iloc[:,[0,1,2,3,4,5,10]]
combination.head(5)
```

```python
# 搭建后面合并其他表格需要用的列

combination['datetime']=pd.to_datetime(combination['trade_date'],format='%Y%m%d')
combination['YYYYMM']=combination['trade_date'].str[0:6]
combination['YYYYQQ']=combination['datetime'].map(lambda x:str(x.year)+'Q'+str(x.quarter))
combination.head(5)
```

```python
# 获取指数历史数据'000001.SH','000005.SH','000006.SH','000300.SH','000905.SH','399001.SZ','399005.SZ','399006.SZ','399016.SZ','399300.SZ','399905.SZ'
pro = ts.pro_api()
df=pd.DataFrame()
for idx in index:
    idx=pro.index_daily(ts_code=idx,start_date=startdate,end_date=enddate,fields='ts_code,trade_date,close,vol')
    df=pd.concat([df,idx],axis=0,ignore_index=True)
df.head(5)
```

```python
# 转换数据格式
df=df.pivot('trade_date','ts_code',['close','vol'])
```

```python
# 将2层列名做出1层的，方便合并
col=[]
for i in df.columns:
    col.append(i[0]+i[1])
df.columns=col
```

```python
df.head(5)
```

```python
# 合并指数
combination=pd.merge(combination,df,how='left',left_on='trade_date',right_on='trade_date',left_index=False,right_index=False,copy=True)
combination.head(5)
```

```python
kpi = "'基本每股收益(元)','每股净资产(元)','每股未分配利润(元)','每股经营现金流(元)','营业总收入(元)','经营现金流/营业收入','毛利润(元)','净利率(%)','总资产周转率(次)','流动比率'"
sql="select code,kpi,period,value from stock_analysis where code = '%s'" % stocklist[0][0:6]
finance = pd.read_sql_query(sql, engine_ts)
finance.head(5)
```

```python
finance=finance[finance['kpi'].isin(['基本每股收益(元)','每股净资产(元)','每股未分配利润(元)','每股经营现金流(元)','营业总收入(元)',
             '经营现金流/营业收入','毛利润(元)','净利率(%)','总资产周转率(次)','流动比率'])]
# finance.pivot(['code','period'],'kpi','value')
finance.head(5)
```

```python
def ws(x):  #  定义数据清洗函数
    if x[-2:] == "万亿":
        r=float(x[:-2]) * 10000*100000000
    elif x[-1] == "万":
        r=float(x[:-1]) * 10000
    elif x[-1] == "亿":
        r=float(x[:-1]) * 100000000
    elif x=='--':
        r=0
    else:
        r=x
    return r
```

```python
finance['value']=finance.value.apply(ws)
```

```python
finance=finance.pivot(index='period',columns='kpi',values='value')
```

```python
finance.reset_index(inplace=True)
```

```python
finance.head(5)
```

```python

finance['period']= pd.to_datetime(finance['period'])
finance.head(5)
```

```python
finance['YYYYQQ']=finance.period.map(lambda x: str(x.year)+'Q'+str(x.quarter))
finance.head(5)
```

```python
# 合并财务数据
combination=pd.merge(combination,finance,how='left',left_on=['YYYYQQ'],right_on=['YYYYQQ'],left_index=False,right_index=False,copy=True)
combination.head(5)
```

```python
combination.shape
```

```python
# 纵向填充，用后面的值填充前面，用以补充最新的数据/数据排序为从最近到以前
combination.fillna(method='bfill',axis=0,inplace=True) 
```

```python
# 纵向填充，用前面的值填充后面面，用以补充最早没有的数据/数据排序为从最近到以前
combination.fillna(method='ffill',axis=0,inplace=True) 
```

```python
# 增加一列计算交易日是星期几，基于假设周几对股票价格有影响
combination['weekday']=combination.datetime.map(lambda x: x.weekday())
```
