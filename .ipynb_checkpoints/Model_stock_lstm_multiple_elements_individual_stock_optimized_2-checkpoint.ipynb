```python
import pandas as pd
import tushare as ts
from sqlalchemy import create_engine
import time
import datetime
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Dropout
import math
from sklearn.metrics import mean_squared_error
```

```python
# 设置参数
asset='E' # E股票 I沪深指数 C数字货币 FT期货 FD基金 O期权 CB可转债（v1.2.39），默认E
ts_code='000001.SZ'
lookfwd=5
lookback=100
split=0.8
cutoff='20201031'
```

```python
# 在线获取历史记录
def get_stock_history_remote(ts_code):
    with open('./parameters.json', 'r') as f:
        p = json.load(f)
        user = p['user']
        port = p['port']
        psw = p['password']
        host = p['host']
        token = p['TU_share_pro_taken']
        cnnstr = "mysql://" + user + ":" + psw + "@" + host + ":" + str(port) + "/stock?charset=utf8&use_unicode=1"
    engine_ts = create_engine(cnnstr)
    ts.set_token(token)
    df = ts.pro_bar(ts_code=ts_code, end_date=cutoff,
                            asset=asset, adj='qfq', freq='D')
    df=df[['ts_code','trade_date','open','high','low','close','amount']]
    df.sort_values(by='trade_date',inplace=True)
    return df
```

```python
df=get_stock_history_remote(ts_code)
df.head(5)
```

```python
df.drop(df[np.isnan(df['close'])].index, inplace=True) 
```

```python
df.head(5)
```

```python
df.drop(['ts_code','trade_date'],axis=1,inplace=True)
df.tail(5)
```

```python
# 计算收盘价未来n天的平均价格
moving_ndays_close=[]
for i in range(0,len(df)-lookfwd):
    ndays=df.iloc[i:i+lookfwd,3]
    moving_ndays_close.append(np.mean(ndays))
moving_ndays_close=moving_ndays_close+list(df['close'][-lookfwd:])  # 直接用收盘价来不足最后5天的移动平均价格
df['moving_ndays_close']=moving_ndays_close
df.tail(10)
```

```python
# 划分训练集、测试集
def data_split(data,split):
    training_size = int(len(data)*split)
    test_size = len(data)-training_size
    traing_data = data.iloc[0:training_size,:]
    test_data = data.iloc[training_size:len(data),:]
    return(traing_data,test_data)

traing_data,test_data=data_split(df,split)
```

```python
traing_data.shape,test_data.shape
```

```python
# 将数据转成0-1之间
scaler = MinMaxScaler(feature_range=(0,1))
scaler.fit(traing_data)

# traing_data = scaler.fit_transform(traing_data)
# traing_data
```

```python
# 构建函数来预处理数据,移动数据框
def create_dataset(dataset, lookback):
    X,Y=[],[]
    for i in range(lookback,len(dataset)):
        a=dataset[(i-lookback):i,:] # 0-99 行  所有列
        b=dataset[i,-1]# 100行 最后一列
        X.append(a)      
        Y.append(b) 
    return np.array(X), np.array(Y)
```

```python
x_train,y_train = create_dataset(traing_data,lookback=lookback)
x_test,y_test = create_dataset(test_data,lookback=lookback)
x_train.shape,y_train.shape
```

```python
#搭建模型
model = Sequential()
model.add(LSTM(32, input_shape=(x_train.shape[1],x_train.shape[2]), return_sequences=True))
# model.add(Dropout(0.2))
model.add(LSTM(16, return_sequences=False))
# model.add(Dropout(0.2))
# model.add(Dense(neurons[2],kernel_initializer="uniform",activation='relu'))        
# model.add(Dense(neurons[3],kernel_initializer="uniform",activation='linear'))
model.add(Dense(1))
# model = load_model('my_LSTM_stock_model1000.h5')
#     adam = keras.optimizers.Adam(decay=0.2)
```

```python
model.compile(loss='mse',optimizer='adam')
```

```python
model.summary()
```

```python
# 训练数据
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=32,verbose=1)
```

```python
# 查看模型收敛状况
plt.plot(history.history['loss'],label='train')
plt.plot(history.history['val_loss'],label='test')
plt.legend()
plt.show()
```

```python
def result_difference(test_predict, real_price):
    test_predict=np.array(test_predict)
    real_price=np.array(real_price)
    diff=[]
    for i in range(len(test_predict)):
        predit_price=round(test_predict[i],2)
        actual_price=round(real_price[i],2)
        price_diff=round(predit_price-actual_price,2)
        percentage_diff=round((predit_price-actual_price)/actual_price,2)
        diff.append([predit_price,actual_price,price_diff,percentage_diff])
    return np.array(diff)
p = result_difference(test_predict[:,-1], df['moving_ndays_close'][-test_predict.shape[0]:])
p
```

```python
plt.figure(figsize=(50,25))
plt.plot(p[:,-1],c='red',marker="o",label="diff %")
plt.title(ts_code)
plt.legend()
plt.show()
```
