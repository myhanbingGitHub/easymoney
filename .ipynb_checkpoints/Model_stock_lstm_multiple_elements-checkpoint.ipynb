```python
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
import math
from sklearn.metrics import mean_squared_error
```

```python
engine = create_engine('mysql://stock:Stock_789@192.168.88.106:3306/stock?charset=utf8&use_unicode=1')
sql = "select * from stock.stock_historic where ts_code='000001.SZ'"
df=pd.read_sql_query(sql,engine)
df.sort_values(by='trade_date',inplace=True,ascending=True)
df.head()
```

```python
df1=df.reset_index()['close']
```

```python
plt.plot(df1)
```

```python
n=np.array(df1).reshape(-1,1) # 将df转成数组，并增加一个维度
n.shape
```

```python
# 将数据转成0-1之间
scaler= MinMaxScaler(feature_range=(0,1))
df1=scaler.fit_transform(n)
df1
```

```python
# 划分训练集、测试集
training_size = int(len(df1)*0.70)
test_size = len(df1)-training_size
print(training_size,test_size)
traing_data=df1[0:training_size,:]
test_data=df1[training_size:len(df1),:1]
test_data
```

```python
# 构建函数来预处理数据
def create_dataset(dataset, days=5):  # 设置days默认为往前看5个交易日
    X,Y=[],[]
    for i in range(len(dataset)-days-1):
        a=dataset[i:(i+days),0]
        X.append(a)      # 假设i=1,days=5 则X=[1,2,3,4,5]
        Y.append(dataset[(i+days,0)]) # 假设i=1,days=5则y=[6]
    return np.array(X), np.array(Y)
```

```python
days= 100  # 定义时间跨度 往前看100天的数据
x_train,y_train = create_dataset(traing_data,days)
x_test,y_test = create_dataset(test_data,days)
x_train.shape,y_train.shape
```

```python
# 将数据维度增加1，用以满足模型输入要求
x_train=x_train.reshape(x_train.shape[0],x_train.shape[1],1) # 3维度
x_test=x_test.reshape(x_test.shape[0],x_test.shape[1],1)  # 3维度
```

```python
# 开始创立模型
model= Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1))) 
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
# 训练数据 输入是2567个样本，每个样本100 x 1 所以结构式 (2567,100,1)

model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=64,verbose=1)
```

```python
# 预测数据
train_predict=model.predict(x_train)
test_predict=model.predict(x_test)
train_predict.shape
```

```python
# 数据还原
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
train_predict.shape
```

```python
# 计算平方差
train_rmse= math.sqrt(mean_squared_error(y_train,train_predict))
test_rmse=math.sqrt(mean_squared_error(y_test,test_predict))
train_rmse,test_rmse
```

```python
# 绘制计算结果图
look_back=100
trainPredictPlot=np.empty_like(df1)
trainPredictPlot[:,:]=np.nan
# 预测值赋给一个列表，位置整体往后移动100
trainPredictPlot[look_back:len(train_predict)+look_back,:]=train_predict
testPredictPlot=np.empty_like(df1)
testPredictPlot[:,:]=np.nan
# 测试值赋给一个列表，位置整体往后移动100*2
testPredictPlot[len(train_predict)+(look_back)*2+1:len(df1)-1,:]=test_predict
plt.figure(figsize=(30,10))
plt.plot(scaler.inverse_transform(df1),c='blue',label="real data")
plt.plot(trainPredictPlot,c='yellow',label="predict per train data")
plt.plot(testPredictPlot,c='red',label="predict per test data")
plt.title('000001.SZ')
plt.legend()
plt.show()
```

```python
# 绘制计算结果图,测试数据部分
real=scaler.inverse_transform(df1)[training_size+100+1:,:]
real.shape,test_predict.shape
plt.figure(figsize=(30,10))
plt.plot(real,c='blue',label="real data")
plt.plot(test_predict,c='red',label="predict per test data")
plt.title('000001.SZ')
plt.legend()
plt.show()
```
